```python
import cv2, numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns, joblib, os
print('Libraries ready')
```

## 1. Font Image Preprocessing

```python
def preprocess_font_image(path, size=(32, 32)):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, size)
    img = img / 255.0
    return img.flatten()

# Synthetic data for demonstration
np.random.seed(42)
FONT_FAMILIES = ['serif', 'sans-serif', 'script', 'monospace', 'display']
n_samples = 400

X = np.random.rand(n_samples, 32*32)  # Flattened 32x32 images
y = np.random.choice(FONT_FAMILIES, n_samples)

print(f'Font dataset: {X.shape[0]} samples, {X.shape[1]} features')
print('Class distribution:', pd.Series(y).value_counts().to_dict())
```

## 2. Font Classification with KNN

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_enc = le.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_enc, test_size=0.2, random_state=42)

# Train KNN
knn = KNeighborsClassifier(n_neighbors=5, metric='cosine', weights='distance')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print('Classification Report:')
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Confusion matrix
fig, ax = plt.subplots(figsize=(7, 5))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', ax=ax, xticklabels=le.classes_, yticklabels=le.classes_, cmap='Blues')
ax.set_title('Font KNN Confusion Matrix'); ax.set_ylabel('True'); ax.set_xlabel('Predicted')
plt.tight_layout(); plt.savefig('font_confusion_matrix.png', dpi=150); plt.show()
```

## 3. Brand Persona → Font Mapping

```python
# Industry to font family lookup table
font_mapping = {
    'Finance':       {'primary': 'serif',      'secondary': 'sans-serif', 'rationale': 'Trust & authority'},
    'Technology':    {'primary': 'sans-serif',  'secondary': 'monospace',  'rationale': 'Modern & clean'},
    'Healthcare':    {'primary': 'sans-serif',  'secondary': 'serif',      'rationale': 'Professional & clear'},
    'Fashion':       {'primary': 'display',     'secondary': 'serif',      'rationale': 'Elegant & editorial'},
    'Education':     {'primary': 'serif',       'secondary': 'sans-serif', 'rationale': 'Authoritative & readable'},
    'Food & Beverage':{'primary': 'script',     'secondary': 'sans-serif', 'rationale': 'Warm & approachable'},
    'Retail':        {'primary': 'sans-serif',  'secondary': 'display',    'rationale': 'Friendly & bold'},
    'Travel':        {'primary': 'display',     'secondary': 'sans-serif', 'rationale': 'Adventurous & inviting'},
    'Sustainability':{'primary': 'sans-serif',  'secondary': 'script',     'rationale': 'Natural & organic'},
}

mapping_df = pd.DataFrame(font_mapping).T.reset_index()
mapping_df.columns = ['Industry', 'Primary Font Family', 'Secondary Font Family', 'Rationale']
print(mapping_df.to_string(index=False))
mapping_df.to_csv('industry_font_mapping.csv', index=False)
print('\nSaved: industry_font_mapping.csv')
```
