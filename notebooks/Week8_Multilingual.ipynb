```python
import json, os, zipfile
import pandas as pd
import google.generativeai as genai
import nltk
nltk.download(['punkt','punkt_tab'], quiet=True)
os.makedirs('outputs/multilingual', exist_ok=True)
print('Ready')
```

## 1. Translation Preparation

```python
GEMINI_API_KEY = 'YOUR_API_KEY_HERE'
genai.configure(api_key=GEMINI_API_KEY)
gemini = genai.GenerativeModel('gemini-pro')

# Assets from previous weeks
assets = {
    'company': 'BrandMind',
    'slogan': 'Intelligence meets identity.',
    'caption': 'Redefining what it means to build a brand. Powered by AI.',
    'hashtags': ['#BrandMind', '#AI', '#Branding', '#Innovation'],
    'tone': 'Professional and innovative',
}

TARGET_LANGUAGES = [
    {'lang': 'Spanish',    'flag': '🇪🇸', 'market': 'Latin America / Spain'},
    {'lang': 'French',     'flag': '🇫🇷', 'market': 'France / West Africa'},
    {'lang': 'German',     'flag': '🇩🇪', 'market': 'Germany / DACH Region'},
    {'lang': 'Japanese',   'flag': '🇯🇵', 'market': 'Japan'},
    {'lang': 'Arabic',     'flag': '🇸🇦', 'market': 'Middle East / North Africa'},
]
print(f'Translating {len(assets)} assets into {len(TARGET_LANGUAGES)} languages')
```

## 2. Multilingual Generation

```python
def translate_content(text, target_lang, tone, content_type='slogan'):
    prompt = f"""Translate the following {content_type} to {target_lang}.
Preserve the {tone} tone, cultural resonance, and brand impact.
Original: '{text}'
Return ONLY the translated text, no explanation."""
    try:
        return gemini.generate_content(prompt).text.strip()
    except Exception as e:
        print(f'  API error for {target_lang}: {e}')
        return f'[{target_lang} translation — configure Gemini API key]'

translations = {}
for lang_info in TARGET_LANGUAGES:
    lang = lang_info['lang']
    print(f"\nTranslating to {lang_info['flag']} {lang}...")
    translations[lang] = {
        'flag':    lang_info['flag'],
        'market':  lang_info['market'],
        'slogan':  translate_content(assets['slogan'],  lang, assets['tone'], 'slogan'),
        'caption': translate_content(assets['caption'], lang, assets['tone'], 'marketing caption'),
    }
    print(f"  Slogan: {translations[lang]['slogan']}")
    print(f"  Caption: {translations[lang]['caption'][:60]}...")
```

## 3. BLEU Score Validation

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import matplotlib.pyplot as plt

smoothie = SmoothingFunction().method1
reference = [assets['slogan'].lower().split()]

bleu_scores = {}
for lang, data in translations.items():
    hypothesis = data['slogan'].lower().split()
    score = sentence_bleu(reference, hypothesis, smoothing_function=smoothie)
    bleu_scores[lang] = round(score, 4)
    print(f"{data['flag']} {lang:10s}: BLEU = {score:.4f}")

# Visualise
fig, ax = plt.subplots(figsize=(10, 4))
langs = list(bleu_scores.keys())
scores = list(bleu_scores.values())
colors = ['#b8975a' if s > 0.1 else '#e8e4db' for s in scores]
ax.bar(langs, scores, color=colors)
ax.axhline(y=0.1, color='#c0392b', linestyle='--', label='Min threshold')
ax.set_ylabel('BLEU Score')
ax.set_title('Translation Quality Validation (BLEU Scores)', fontweight='bold')
ax.legend()
ax.spines[['top','right']].set_visible(False)
plt.tight_layout(); plt.savefig('outputs/bleu_scores.png', dpi=150); plt.show()
```

## 4. Language-Specific Campaign ZIPs

```python
for lang, data in translations.items():
    lang_dir = f"outputs/multilingual/{lang.lower()}"
    os.makedirs(lang_dir, exist_ok=True)

    with open(f"{lang_dir}/campaign_copy.txt", 'w', encoding='utf-8') as f:
        f.write(f"{data['flag']} {lang} Campaign Copy\n")
        f.write('='*40 + '\n\n')
        f.write(f"Slogan: {data['slogan']}\n\n")
        f.write(f"Caption: {data['caption']}\n\n")
        f.write(f"Hashtags: {' '.join(assets['hashtags'])}\n")
        f.write(f"Target Market: {data['market']}\n")

    # Per-language ZIP
    zip_path = f"outputs/multilingual/{lang.lower()}_campaign.zip"
    with zipfile.ZipFile(zip_path, 'w') as zf:
        zf.write(f"{lang_dir}/campaign_copy.txt", 'campaign_copy.txt')
    print(f"Created: {zip_path}")

# Save full translations JSON
with open('outputs/multilingual/all_translations.json', 'w', encoding='utf-8') as f:
    json.dump({'source': assets, 'translations': translations, 'bleu_scores': bleu_scores}, f, indent=2, ensure_ascii=False)
print('\nSaved: outputs/multilingual/all_translations.json')
```
